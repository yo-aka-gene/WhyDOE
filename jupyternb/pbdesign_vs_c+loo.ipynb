```python
import matplotlib.pyplot as plt

from doe_modules.design import CLOO, PlackettBurman
from doe_modules.plot import correlation_heatmap, design_heatmap
from doe_modules.preferences import kwarg_savefig, outputdir
```

```python
for name, doe in {"C+LOO": CLOO, "Plackett-Burman": PlackettBurman}.items():

    fig, ax = plt.subplots(figsize=(3, 3))

    design_heatmap(
        design=doe, n_factor=8, ax=ax,
        cmap="Purples", square=True,
        annot_kws={"+": 1, "-": -1}
    )
    ax.set(title=f"{name} Design")
    fig.savefig(f"{outputdir}/{name.lower()}.png", **kwarg_savefig)
```

```python
for name, doe in {"C+LOO": CLOO, "Plackett-Burman": PlackettBurman}.items():

    fig, ax = plt.subplots(figsize=(10, 8))

    correlation_heatmap(
        design=doe, n_factor=8, ax=ax,
        cmap="coolwarm", square=True, 
    )
    ax.set(title=f"{name} Design")
    fig.savefig(f"{outputdir}/{name.lower()}_corr.png", **kwarg_savefig)
```
